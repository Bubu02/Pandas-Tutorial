```python
 # Creating seperate dataframe for all independent variables
independent_variables = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']
x = heart[independent_variables]

print(type(x))
print(x)

target_variables = 'target'
y = heart[target_variables]

print(type(y))
print(y)
```

```python
from sklearn.preprocessing import MinMaxScaler
# import tensorflow as tf
# from tensorflow import keras
# from keras.models import Sequential
# from keras.layers import Dense

def getNormalizationData(x):

    # Fit scaler on training data
    norm = MinMaxScaler().fit(x)

    # transform training data
    x_norm = norm.transform(x)
    x_train_norm = pd.DataFrame(x_norm, columns= x.columns.values)
    
    return x_train_norm;
```

```python
# Normalization typically means rescales the values into a range of [0,1]
x_train_norm = getNormalizationData(x);
x_train_norm.head()
```

```python
x_train_norm.describe()
```

```python
# Convert to numpy array
x_numpy = np.array(x_train_norm)
y_numpy = np.array(y)

print(type(x_numpy))
print(type(y_numpy))
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_numpy, y_numpy, test_size = 0.20, random_state=42)
```

## Model Selection

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier
# model = LinearRegression()
model = RandomForestClassifier()
# model = RandomForestRegressor()
model.fit(x_train, y_train)
```

```python
print("Making predictions for the following in train set:")
predicted_values = model.predict(x_train[0:10,:])
true_values = y_train[0:10]
print("True values are ", true_values)
print("The predicted values are", predicted_values)
print("Diff", abs(predicted_values-true_values))
```

```python
from sklearn.metrics import accuracy_score
# Average training error
dfTest = y_train-model.predict(x_train)
print("average Train error ", np.absolute(dfTest).mean())

# Accuracy Score in Testing set
testing_accuracy = accuracy_score(true_values, predicted_values, normalize=False)
print("Accuracy Score in Testing set", testing_accuracy)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix

# Assuming y_train and y_test are arrays/lists of discrete class labels
# Ensure that rf_model is a trained classifier
# x_train and x_test should be your feature data

# Plot the confusion matrix for training data including numbers in the plot
cm_train = confusion_matrix(y_train, model.predict(x_train))
ConfusionMatrixDisplay(confusion_matrix=cm_train, display_labels=model.classes_).plot(cmap='gist_earth', values_format='d')
plt.title("Confusion Matrix (Training)")

# Plot the confusion matrix for testing data, include numbers in the plot
cm_test = confusion_matrix(y_test, model.predict(x_test))
ConfusionMatrixDisplay(confusion_matrix=cm_test, display_labels=model.classes_).plot(cmap='gist_earth', values_format='d')
plt.title("Confusion Matrix (Testing)")

plt.show()
```

```python
# average Training accuracy
training_accuracy = accuracy_score(y_train, model.predict(x_train))
print("Average Training accuracy ", training_accuracy)

# average Testing Accuracy
testing_accuracy = accuracy_score(y_test, model.predict(x_test))
print("Average Testing Accuracy", testing_accuracy)
```

```python
from sklearn.metrics import  precision_score, recall_score, f1_score, precision_recall_curve

precision = precision_score(y_test, model.predict(x_test), average='weighted')
print("Precision:", precision)
```

```python
recall = recall_score(y_test, model.predict(x_test), average='weighted')
print("Recall:", recall)
```

```python
F1 = f1_score(y_test, model.predict(x_test), average='weighted')
print("F1:", F1)
```

```python
# Precesion recall curve
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt

# Predict probabilities
y_scores = model.predict_proba(x_test)

# Compute Precision-Recall and plot curve
precision, recall, thresholds = precision_recall_curve(y_test, y_scores[:, 1])

plt.figure()
plt.plot(recall, precision, label='Precision-Recall curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.title('Precision-Recall Curve')
plt.legend(loc="lower left")
plt.grid()
plt.show()
```
